Google Calendar Chart

[![Binder](https://mybinder.org/badge_logo.svg)](https://mybinder.org/v2/gh/fslaborg/XPlot/gh-pages?filepath=google-calendar-chart.ipynb)



```
open XPlot.GoogleCharts
open System

let data =
    let rnd = Random()
    [
        for x in 1. .. 600. ->
            DateTime(2013, 1, 9).AddDays(x), rnd.Next(0, 5)
    ]

let options =
    Options(
        title = "GitHub Contributions",
        height = 350
    )

let chart =
    data
    |> Chart.Calendar
    |> Chart.WithOptions options
```
